### Defined the Class for getting the State of Graph at every nodee

```python
from typing_extensions import TypedDict
class State(TypedDict):
    graph_info: str
```

### Nodes

```python
def start_play(state:State):
    print("Start Play node has been called")
    return {"graph_info": state["graph_info"] + "I am planning to play "}

def cricket(state:State):
    print("Cricket node has been called ")
    return {"graph_info": state["graph_info"] + "Cricket"}

def bedminton(state:State):
    print("Bedminton node has been called")
    return{"graph_info":state["graph_info"] + "Bedminton"}
```

```python
import random
from typing import Literal # Constant

def random_play(state:State)-> Literal["cricket","bedminton"]:
    if random.random() > 0.5:
        return "cricket"
    else:
        return "bedminton"
```

```python
from IPython.display import Image,display
from langgraph.graph import StateGraph,START, END

##Building Graph
graph = StateGraph(State)

## Add all the nodes

graph.add_node("start_play",start_play)
graph.add_node("cricket",cricket)
graph.add_node("bedminton",bedminton)

#Schedule the flow of the graph
graph.add_edge(START,"start_play")
graph.add_conditional_edges("start_play",random_play)
graph.add_edge("cricket",END)
graph.add_edge("bedminton",END)

### Compile the graph

graph_builder = graph.compile()

```

```python
graph_builder.invoke({"graph_info":"My name is Shaheer "})
```
